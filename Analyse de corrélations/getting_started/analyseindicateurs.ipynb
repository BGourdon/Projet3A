# Étude de l'évolution des indicateurs (en fonction des années, des scénarios et du niveau d'étude)

```r
source("manipdatasets.R")
```

### *Manipulation des différents datasets pour le fichier MAELIA_crop*

```r
ScenarioFerme <- fichierfarm("Baseline situation", "AF1", "Yield")
```

```r
qplot(Year, Yield, data = ScenarioFerme) + facet_wrap(~ Crop)
```

```r
ggplot(ScenarioFerme) + 
  geom_line(aes(x = Year, y = Yield, color = Crop)) 
```

```r
ScenarioCulture <- fichierculture("Baseline situation", "WheatW", "Yield")
```

```r
qplot(Year, Yield, data = ScenarioCulture) + facet_wrap( ~ Farm)
```

```r
ggplot(ScenarioCulture) + geom_boxplot(aes(x = Farm, y = Yield))
```

```r
ggplot(ScenarioCulture) + 
  geom_line(aes(x = Year, y = Yield, color = Farm)) 
```

```r
ScenarioIndicateur <- fichierindicateur("Coexistence scenario", "Revenue")
```

```r
ggplot(ScenarioIndicateur) + 
  geom_line(aes(x = Year, y = Revenue, color = Crop)) + facet_wrap( ~ Farm, scales = "free_y")
```

```r
indicateurAFi <- fichierscenario("AF2", "Revenue")
```

```r
ggplot(indicateurAFi, aes(x=Year)) + 
  geom_line(aes(x = Year, y = Revenue, color = Crop)) + facet_wrap( ~ Scenario)
```

```r
IndicateurCulture <- fichierscenario2("WheatW", "Yield")
```

```r
ggplot(IndicateurCulture, aes(x=Year)) + 
  geom_line(aes(x = Year, y = Yield, color = Farm)) + facet_wrap( ~ Scenario)
```

```r
ggplot(IndicateurCulture, aes(x=Year)) +
  geom_boxplot(aes(x = Farm, y = Yield)) + facet_wrap( ~ Scenario)
```
